```python
import os
```

```python
path = "racer"
```

```python
folders = ["images", "psnr", "ssim"]
```

```python
for folder in folders:
    for subfolder in os.listdir(folder):
        if subfolder != ".DS_Store":
            for i in os.listdir(f"{folder}/{subfolder}"):
                if i != ".DS_Store":
                    os.system(f"ffmpeg -i {folder}/{subfolder}/{i} {folder}/{subfolder}/{i.split('.')[0]}.webp")
                    os.system(f"rm {folder}/{subfolder}/{i}")
```

```python
for i in os.listdir(path):
    os.system(f"ffmpeg -i {path}/{i} {path}/{i.split('.')[0]}.webp")
```
